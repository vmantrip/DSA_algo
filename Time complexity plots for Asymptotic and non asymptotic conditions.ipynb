```python
import numpy as np
import matplotlib.pyplot as plt
x = [(i+1)*10 for i in range(0,11)]
y1 = [i-1 for i in x]
y2 = [np.log(i) for i in x]
y3 = [i/2 for i in x]
y4 = [i**0.5 for i in x]
y5 = [i**(1/3) for i in x]
y6 = [i/np.log(i) for i in x]
```

```python
#Asymptotic assumption plots
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.plot(x, y4)
plt.plot(x, y5)
plt.plot(x, y6)
plt.show()
```

```python
x = [(i+2) for i in range(0,9)]
y1 = [i-1 for i in x]
y2 = [np.log(i) for i in x]
y3 = [i/2 for i in x]
y4 = [i**0.5 for i in x]
y5 = [i**(1/3) for i in x]
y6 = [i/np.log(i) for i in x]
```

```python
# Plots where n < n0
# n0 is the minimum number of observations required to achieve asymptotic asssumption
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.plot(x, y4)
plt.plot(x, y5)
plt.plot(x, y6)
plt.show()
```

## Observations:
Based on the 2 graphs above, we can conclude that value of n0 that Asymptotic condition becomes valid is n0=10

We can understand that time complexity behavoiur changes with the number of observations involved in our analysis.
Hence we must be clear with our idea of what is small and what is big.
